## Feature Engineering
Ingenieria de características/variables

```python
#housing_df["categorias_ingresos"] = pd.cut(housing_df["median_income"], bins=5, labels=[1, 2, 3, 4, 5])
```

Para comentar todo un bloque: ctrl + /

```python
housing_df["categorias_ingresos"] = np.ceil(housing_df["median_income"]/1.5)
housing_df["categorias_ingresos"] = housing_df["categorias_ingresos"].apply(lambda x: 5 if x>5 else x)
housing_df["categorias_ingresos"]
```

```python
housing_df["categorias_ingresos"].value_counts()
```

```python
# Analizaremos la distribucion de categorias_ingresos
sns.displot(housing_df, x='categorias_ingresos', bins=30, color="m")
```
